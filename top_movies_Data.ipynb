```python

#importing all necessary libraries for data manupilating(pandas) and datavisualisation(seaborn, matplotlib)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading our csv file and printing it
movie = pd.read_csv('imdbTop250.csv')
movie
```

```python
# Checking last 5 items in our data frame
movie.tail() 
```

### Create a histogram on 'Date' column as shown below

```python
#Creating histogram of "date" column which is showing how many of those movies appeard which year
sns.histplot(movie.Date, binwidth = 3)
```

```python
movie[movie.Title == "The Shawshank Redemption"]

#just deleting one column
movie.drop("IMDBlink", axis=1, inplace=True)
movie
```

```python
#Creating new column where we will keep only first genre(some movies have several of them)
def first_genre(genres): #Animation, Short, Comedy
    splitted_list = genres.split(", ")
    return splitted_list[0]
movie['Genre'] = movie["Genre"].apply(first_genre)
movie["Genre"].tail(20)
```

```python
#getting info about columns types and amount of items in there
movie.info()
```

```python
#Finding out whic director appear most in that data frame
kekus = movie.value_counts("Director")
kekus
```

movie.info()

```python
#How many movies do we have in each genre
how_many_genres = movie.value_counts("Genre")
how_many_genres
```

```python
#Let's visualize how many movies do we have in each genre

plt.xticks(rotation=45);

ax = sns.countplot(x="Genre", data = movie)
#🦍🦍🦍
```

```python
#score oT new_genre 
#Using our new_genre column(which contains only one genre for one movie) let's see which genre is the most popular
figure, ax = plt.subplots(figsize = (25, 8))
sns.barplot(x = "Genre", y = "Score", data = movie, ax = ax)
```

```python
#let's if there is any correlation between year of movie and it's score, using scatterplot
#( there is, as we can see)
figure, ax = plt.subplots(figsize = (8, 5 ))
sns.scatterplot(x = "Date", y = "Score", data = movie, ax = ax)
```

```python
# Let's find out if there is relation between Run Time and Score of movie
#If you want to make a mvoie, make sure duration of that movie is in between 90 and 150 minutes
figure, ax = plt.subplots(figsize = (7, 5))
sns.scatterplot(x = "RunTime", y = "Score", data = movie, ax = ax)
```
